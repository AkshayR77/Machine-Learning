```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/sample_data/Iris.csv")
df
```

```python
df=df.drop(['Id'],axis=1)
df
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_test
```

```python
#standerd scaler equation

# z=(x-u)/s

from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df1=pd.DataFrame({'Actual_Value':y_test,'Predicted_Value':y_pred})
df1
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,ConfusionMatrixDisplay
result=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
score
```

```python
#plot of decision tree
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
tree.plot_tree(model,feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],class_names=['Iris-setosa','Iris-versicolor','Iris-virginica'],filled=True)
```
